```python
from ragas.testset.generator import TestsetGenerator
from ragas.testset.evolutions import simple, reasoning, multi_context, conditional
from ragas.llms import LangchainLLMWrapper
from ragas.embeddings import LangchainEmbeddingsWrapper
from ragas.testset.extractor import KeyphraseExtractor
from ragas.testset.docstore import InMemoryDocumentStore
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
#ë°ì´í„° ìƒì„±ê¸°
generator_llm = ChatOpenAI(model='gpt-4o')
```

```python
# ë¹„í‰ê¸°
critic_llm = ChatOpenAI(model='gpt-4o')

# ìƒì„±ê³¼ ë¹„í‰ì˜ ëª¨ë¸ì„ ê°™ê²Œ í•˜ë¯€ë¡œ ì¼ê´€ì„±
```

```python
# ì„ë² ë”© ëª¨ë¸
embeddings = OpenAIEmbeddings(model = 'text-embedding-3-small')
```

```python
# í…ìŠ¤íŠ¸ ë¶„í• ê¸° ì„¤ì •
splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap=100)
```

```python
langchain_llm = LangchainLLMWrapper(ChatOpenAI(model='gpt-4o'))
# êµ¬ë¬¸ì¶”ì¶œê¸° ìƒì„±ì„ ìœ„í•œ ëª¨ë¸ í˜¸ì¶œ(RAGASì™€ì˜ í˜¸í™˜ì„ ìœ„í•œ wrapper)
# ëŒ€ëŸ‰ì˜ í…ìŠ¤íŠ¸ë¥¼ ì²˜ë¦¬ í•  ë•Œ í•µì‹¬ì„ ë‹´ëŠ”, í‚¤ì›Œë“œ ì¤‘ì‹¬ìœ¼ë¡œ ì§„í–‰ í•  ë•Œ 
```

```python
# êµ¬ë¬¸ì¶”ì¶œê¸° : ë¬¸ì„œì˜ í•µì‹¬ ì •ë³´ ì‹ë³„ ë° ì¶”ì¶œ ì—­í™œ
Keyphrase_extractor = KeyphraseExtractor(llm = langchain_llm)
```

```python
# ì„ë² ë”© ëª¨ë¸ë¡œ RAGASì™€ì˜ í˜¸í™˜ì„ ìœ„í•´ wrapper ì ìš©
ragas_embeddings = LangchainEmbeddingsWrapper(embeddings)
```

```python
# docstore ë§Œë“¤ê¸°
docstore = InMemoryDocumentStore(
    splitter=splitter,
    embeddings = ragas_embeddings,
    extractor=Keyphrase_extractor
)
```

```python
# ìƒì„±ê¸° ë§Œë“¤ê¸°
generator = TestsetGenerator.from_langchain(
    # ìƒì„±ì„ í•˜ë©´ì„œ í‰ê°€ë¥¼ í•¨ê»˜í•¨
    generator_llm, # ìƒì„±ê¸°
    critic_llm, # íŒë³„ê¸°
    ragas_embeddings, # ì„ë² ë”©ëª¨ë¸
    docstore=docstore # ë¬¸ì„œì €ì¥ì†Œ
    # docstore ëŠ” ìƒëµ ë¶ˆê°€
)
```

```python
# ìƒì„±ê¸°ì— ë‹¤ì–‘í•œ ìœ í˜•ì„ ì•Œë ¤ì¤Œ => ì¡°ê±´ì— ë§ëŠ” ì§ˆë¬¸ ìƒì„±
# ì§ˆë¬¸ ìœ í˜• ë¶„í¬ ê²°ì •
distributions = {simple : 0.4, reasoning:0.2, multi_context:0.2 , conditional:0.2}
# ê°„ë‹¨í•œ ì§ˆë¬¸ 40% , ì¶”ë¡ ì´ í•„ìš”í•œ ì§ˆë¬¸, ì—¬ëŸ¬ê°€ì§€ ë§¥ë½ì„ ê³ ë ¤í•œ ì§ˆë¬¸, ì¡°ê±´ë¬¸ ì§ˆë¬¸ ê° 20%
```

```python
testset = generator.generate_with_langchain_docs(
    documents=docs,
    test_size=10,
    distributions=distributions,
    with_debugging_logs=True,
    # ë¡œê·¸ í™œì„±í™”
    raise_exceptions = False
    # ì˜ˆì™¸ê°€ ë°œìƒí•´ë„ ê³„ì† ì§„í–‰, ì˜ˆì™¸ ë°œìƒ ì‹œ ë¬¸ì„œê°€ ì²˜ë¦¬ ë˜ì§€ ì•Šê±°ë‚˜ ë¬¸ì„œì— ê¸°ë¡ë¨
)
```

```python
import pandas as pd
df = pd.read_csv('./ragas_dataset.csv')
df
```

```python
from datasets import Dataset
test_dataset = Dataset.from_pandas(df)
test_dataset
```

```python
# context ìƒì„±ëœ ì§ˆë¬¸ (ë¬¸ìì—´ => ë¦¬ìŠ¤íŠ¸) í˜•íƒœ ë³€ê²½í•´ì¤˜ì•¼í•¨
df['contexts'][0]
# "[]" ë¦¬ìŠ¤íŠ¸ë¡œ ë³´ì¼ ìˆ˜ ìˆì§€ë§Œ ë¬¸ìì—´ì„
```

```python
import ast
def convert_to_lsit(example):
    contexts = ast.literal_eval(example['contexts'])
    return {'contexts' : contexts}
```

```python
test_dataset = test_dataset.map(convert_to_lsit)
```

```python
test_dataset['contexts'][0]
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_community.vectorstores import FAISS
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
```

```python
loader = PDFPlumberLoader('./data/SPRi AI Brief_11ì›”í˜¸_ì‚°ì—…ë™í–¥_F.pdf')
docs = loader.load()
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap = 50)
split_documenrs = text_splitter.split_documents(docs)
```

```python
empbeddins = OpenAIEmbeddings()
```

```python
vectorstore = FAISS.from_documents(documents = split_documenrs, embedding = embeddings)
```

```python
retriever = vectorstore.as_retriever()
```

```python
prompt = PromptTemplate.from_template(
    """
    ë„ˆëŠ” ì£¼ì–´ì§„ ì§ˆë¬¸ì— ëŒ€ë‹µí•˜ëŠ” AIì•¼. ë‹¤ìŒ ê²€ìƒ‰ëœ contextë¥¼ ì‚¬ìš©í•´ì„œ questionì— ëŒ€ë‹µí•´ì¤˜.
    ë‹µì„ ëª¨ë¥´ë©´, 'ì•Œ ìˆ˜ ì—†ìŠµë‹ˆë‹¤'ë¼ê³  ëŒ€ë‹µí•´

    # Context : {context}
    # Question : {question}
    # Answer :
    """
)
```

```python
llm = ChatOpenAI(model = 'gpt-4o', temperature=0)
```

```python
chain =(
    {'context':retriever,'question':RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
batch_dataset = []
for question in test_dataset['question']:
    batch_dataset.append(question)
```

```python
batch_dataset[:3]
```

```python
answer = chain.batch(batch_dataset)
answer[:3]
```

```python
if 'answer' in test_dataset.column_names:
    test_dataset = test_dataset.remove_columns(['answer']).add_column('answer', answer)
else:
    test_dataset = test_dataset.add_column('answer',answer)
```

```python
test_dataset
```

```python
from ragas import evaluate
from ragas.metrics import answer_relevancy,faithfulness,context_recall,context_precision
```

```python
result = evaluate(
    dataset = test_dataset,
    metrics = [
        context_recall,   # ëª¨ë¸ì´ ë¬¸ë§¥ì—ì„œ ì¤‘ìš”í•œ ì •ë³´ë¥¼ ì˜ íšŒìƒ, ì¬í˜„í–ˆëŠ”ê°€
        faithfulness,     # ëª¨ë¸ì˜ ë‹µë³€ì´ ë¬¸ë§¥ì— ê¸°ë°˜í•´ì„œ ì‚¬ì‹¤ì ì¸ê°€ (factual)
        answer_relevancy, # ì „ì²´ ì§ˆë¬¸ì— ëŒ€í•´ ëª¨ë¸ì˜ ë‹µë³€ì´ ì§ˆë¬¸ê³¼ ì–¼ë§ˆë‚˜ ê´€ë ¨ì„±ì´ ìˆëŠ”ê°€
        context_precision # ëª¨ë¸ì´ ë¬¸ë§¥ì—ì„œ í•„ìš”í•œ ì •ë³´ë¥¼ ì •í™•íˆ í™œìš©í–ˆëŠ”ê°€
    ]
)
```
